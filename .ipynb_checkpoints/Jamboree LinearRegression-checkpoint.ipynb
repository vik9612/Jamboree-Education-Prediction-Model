 ## Multicollinearity check by VIF score

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    
    # Calculating VIF for each feature
    vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                       for i in range(len(X.columns))]
    return round(vif_data,3)

vif_data = calculate_vif(X)
print(vif_data)

# Dropping variables with VIF > 5 one by one and recalculating VIF each time
while vif_data[vif_data['VIF'] > 5].shape[0] > 0:
    highest_vif = vif_data[vif_data['VIF'] == vif_data['VIF'].max()]
    print(f"Dropping: {highest_vif['feature'].values[0]} with VIF: {highest_vif['VIF'].values[0]}")
    X = X.drop(columns=[highest_vif['feature'].values[0]])
    vif_data = calculate_vif(X)
    print(round(vif_data,2))
```

## Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculation of Metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1-r2) * (len(y)-1)/(len(y)-X.shape[1]-1)

print(f"MAE: {round(mae,2)}")
print(f"RMSE: {round(rmse,2)}")
print(f"R2 Score: {round(r2,2)}")
print(f"Adjusted R2 Score: {round(adj_r2,2)}")
```
